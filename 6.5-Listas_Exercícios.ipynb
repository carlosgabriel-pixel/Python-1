### Q5.

Escreva um programa para encontrar a diferença entre duas listas, incluindo elementos duplicados.

Ex:
```
Listas originais: 
[1, 1, 2, 3, 3, 4, 4, 5, 6, 7]
[1, 1, 2, 4, 5, 6]
Diferença entre as listas:
[3, 3, 4, 7]
```

````python
# Implemente sua solução aqui
# Listas originais do exemplo
lista1 = [1, 1, 2, 3, 3, 4, 4, 5, 6, 7]
lista2 = [1, 1, 2, 4, 5, 6]

# Criamos uma cópia da lista2 para não modificar a original
temp_lista2 = lista2.copy()
diferenca = []

# Para cada elemento na primeira lista
for elemento in lista1:
    if elemento in temp_lista2:
        # Se o elemento está na lista2, nós o "consumimos" (removemos uma ocorrência)
        temp_lista2.remove(elemento)
    else:
        # Se não está, ele faz parte da diferença
        diferenca.append(elemento)

print("Diferença entre as listas:")
print(diferenca)
````

```
Tamanho para divisão: 0
```
